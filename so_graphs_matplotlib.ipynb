```python
from matplotlib.pyplot import figure, show
from numpy import arange, sin, cos, pi, e
%matplotlib inline

t1 = arange(0.0, 1.0, 0.01)
t2 = arange(-3.0,3.0,1.0)

fig = figure(1)

ax1 = fig.add_subplot(211)
ax1.plot(t1, sin(2*pi*t1), color="r")
ax1.plot(t1, cos(5*pi*t1), color="b")
ax1.grid(True)
ax1.set_ylim((-2, 2))
ax1.set_ylabel('y')
ax1.set_xlabel('x')
ax1.set_title('y = sin(5x) und y = cos(5x)')

ax2 = fig.add_subplot(212)
ax2.plot(t2, sin(e**t2))

ax2.set_ylim((-2, 2))
ax2.set_xlabel('x')
ax2.set_ylabel('y')

show()
```
