```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time

url = "http://www.ign.com/"
driver = webdriver.Firefox()
driver.get(url)
time.sleep(3)
driver.find_element_by_css_selector('.all-games').click()
time.sleep(3)
html = driver.page_source.encode('utf-8')

soup = BeautifulSoup(html, 'lxml')
games = soup.find_all('article', attrs={"class":"feed-item basic article"})
```
